  3. Use adequadamente paletas pré-definidas
  4. Observe sempre o contexto e objetivo da visualização
  5. Considere as emoções transmitidas pela cor com tranquilidade, entusiasmos, seriedade e muito mais.

```python
# pesquisar os top 10 paises com imigrantes para o canada
top_10 = df.sort_values(by='Total',ascending=False).head(10)
```

```python
top_10
```

```python
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
ax = sns.barplot(data=top_10,y=top_10.index,x='Total',orient='h')
ax.set(
    title='Os dez países com mais imigrantes para o Canadá',
    xlabel='Total de Imigrantes',
    ylabel=''
)
plt.show
```

```python
def gerar_grafico_paleta(pallete):
  fig,ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10,y=top_10.index,x='Total',
                   orient='h',
                   palette=pallete)
  ax.set_title('Os dez países com mais imigrantes para o Canadá',
               loc='left',
               fontsize=16)
  ax.set_xlabel('Total de Imigrantes', fontsize=14)
  ax.set_ylabel('')
  plt.show()
```

# Gráficos Interativos com plotly

```python
import plotly.express as px
```

```python
anos=list(map(str,range(1980,2014)))
```

```python
brasil = df.loc['Brasil',anos]
```

```python
brasil_dict = {'ano':brasil.index.tolist(),'imigrantes':brasil.values.tolist()}
```

```python
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil.head()
```

```python
from matplotlib import pyplot as plt
dados_brasil['imigrantes'].plot(kind='line', figsize=(8, 4), title='imigrantes')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
fig = px.line(dados_brasil, x='ano', y='imigrantes', title='Imigrantes no Brasil')
fig.show()
```

```python
fig = px.line(dados_brasil,
              x='ano',
              y='imigrantes',
              title='Imigrantes do Brasil para o Canada de 1980 a 2013')
# centralizar o título
fig.update_layout(title_x=0.5)
fig.update_layout(width=1000,
                  height=500,
                  xaxis={'tickangle':-45})
fig.update_traces(line_color='red',line_width=3.8)
#fig.update_xaxes(tickangle=-45)
fig.show()
```

# gerando o grafico acima em html

```python
fig.write_html('imigrantes_brasil_tarde.html')
```
